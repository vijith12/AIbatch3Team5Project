```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
%matplotlib inline
```

```python
dataset=pd.read_csv('featuresdf.csv')
```

```python
dataset
```

```python
dataset.groupby("name")['popularity'].mean().sort_values(ascending=False).head()
```

```python
dataset.groupby("name")['popularity'].count().sort_values(ascending=False).head()
```

```python
popularity=pd.DataFrame(dataset.groupby("name")['popularity'].mean())
popularity
```

```python
popularity["popularity count"]=pd.DataFrame(dataset.groupby("name")['popularity'].count())
popularity
```

```python
plt.figure(figsize=(10,5))

popularity["popularity count"].hist(bins=70)
```

```python
plt.figure(figsize=(10,5))

popularity["popularity"].hist(bins=70)
```

```python
sns.jointplot(x="popularity",y="popularity count",data=popularity)
```

```python
songmat=dataset.pivot_table(index="release_year",columns="name",values="popularity")
songmat.head(20)
```

```python
popularity.sort_values("popularity count",ascending=False).head(10)
```

```python
st_populaarity=songmat["Summertime"]
Baby_populaarity=songmat["Baby"]
st_populaarity.head(40)
```

```python
Baby_populaarity.head(10)
```

```python
similarity_st=songmat.corrwith(st_populaarity)
similarity_Baby=songmat.corrwith(Baby_populaarity)
```

```python
corr_st=pd.DataFrame(similarity_st,columns=['correlation'])
corr_st.dropna(inplace=True)
corr_st.sort_values("correlation",ascending=False).head(50)
```

```python
corr_st=corr_st.join(popularity["popularity count"])
```

```python
print(corr_st)
```

```python
corr_st[corr_st["popularity count"]>65].sort_values("correlation",ascending=False).head(20)
```

```python
corr_baby=pd.DataFrame(similarity_Baby,columns=['correlation'])
corr_baby.dropna(inplace=True)
corr_baby=corr_baby.join(popularity["popularity count"])
corr_baby[corr_baby["popularity count"]>5].sort_values("correlation",ascending=False)
```
